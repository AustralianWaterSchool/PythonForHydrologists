
# Animated graphs

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib notebook
```

```python
from scipy.special import exp1

def theis(r, t, S, T, Q=100):
    u = S * r ** 2 / (4.0 * T * t)
    s = Q / (4 * np.pi * T) * exp1(u)
    return s

S = .01
T = 10
r = np.linspace(0.1, 100, 200)
for t in [0.1, 1, 10]:
    plt.plot(r, theis(r, t, S, T), label = f'time = {t} d')

plt.xlabel('radial distance')
plt.ylabel('head ')
plt.legend(loc='best')
```

```python


r = np.linspace(0.1, 1000, 200)

fig, ax = plt.subplots()
ln, = ax.plot(r, np.zeros_like(r))
t_str = ax.text(800, 8, "t = {:0.1f} d".format(0),
                ha='center', va='center')

def init():
    ax.set_xlim(0, 1000)
    ax.set_ylim(10, 0)
    ax.grid(True, ls=':')

def update(t):
    ln.set_ydata(theis(r, t, S, T))
    t_str.set_text("t = {:0.2f} d".format(t))

ani = FuncAnimation(fig, update, frames=[0.1, 1, 10], init_func=init)
```

```python
from matplotlib.animation import FuncAnimation
r = np.linspace(0.1, 1000, 200)

fig, ax = plt.subplots()
ln, = ax.plot(r, np.zeros_like(r))
t_str = ax.text(800, 8, "t = {:0.1f} d".format(0),
                ha='center', va='center')

def init():
    ax.set_xlim(0, 1000)
    ax.set_ylim(10, 0)
    ax.grid(True, ls=':')

def update(t):
    ln.set_ydata(theis(r, t, S, T))
    t_str.set_text("t = {:0.2f} d".format(t))

ani = FuncAnimation(fig, update, frames=np.logspace(-1, 2, 1000), init_func=init,
                    blit=True, interval=2, repeat=True)
```
